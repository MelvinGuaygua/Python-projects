## Plot (grafico)

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.array([1,2,3])  # me doy cuenta de que los arrays debes tener la misma cantidad de numeros
b = np.array([10,12,20]) 
print(a)
print(b)

plt.plot(a,b) # me doy cuenta de que lo que le indico con los arrays son las coordenadas x y 
plt.show()
```

```python
# marker es c√≥mo pintar los puntos
# linestyle es el estilo de la linea
# linewidth es el grosor de la linea
# label es la leyenda
plt.plot(a,b,marker="o", linestyle=":", linewidth="2", color="k", label='8*x') 

plt.legend() #esto es para mostrar el label
plt.show()
```

```python
plt.plot(a,b,marker="o", linestyle=":", linewidth="2", color="r", label='8*x')
plt.plot(a,a*8,marker = '+', linestyle = '--',linewidth = '1.5',color='b',label='8*x')

font = {'family':'serif','color':'darkred','size':15}
plt.title("mi grafico", loc="center", fontdict=font)
plt.legend()
plt.xlabel("x") #descripcion eje x
plt.ylabel("y = $x^2$") #descripcion eje y
plt.grid() #para poner cuadros
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
plt.show()

fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y)
axs[1].plot(x, -y)
plt.show()
```

```python
data = np.load('medida_H18.npz')
M_hombres=data['x']
data = np.load('medida_M18.npz')
M_mujeres=data['x']
print("se han cargado",len(M_hombres), "medidas de hombres",len(M_mujeres), "medidas de mujeres")
```

```python
import numpy as np

mediaHombre = np.mean(M_hombres)
plt.hist(M_hombres, bins=20)
font = {'family':'serif','color':'blue','size':15}
plt.title("Melvin Guaygua", loc="center", fontdict=font)
plt.axvspan(165, 185, color='blue', alpha=0.3) #ESTO ES PARA PONER LA SOMBRA
plt.vlines(x=mediaHombre, ymin=0, ymax=600, linestyles ="solid", colors ="cyan") #ESTO ES PARA PONER LA LINEA VERTICAL 
plt.show()


mediaMujeres = np.mean(M_mujeres)
plt.vlines(x=mediaMujeres, ymin=0, ymax=600, linestyles ="solid", colors ="cyan") #ESTO ES PARA PONER LA LINEA VERTICAL 
plt.vlines(x=mediaHombre, ymin=0, ymax=600, linestyles ="solid", colors ="cyan") #ESTO ES PARA PONER LA LINEA VERTICAL 
plt.hist(M_hombres, bins=20, label="media hombre>18")
plt.axvspan(165, 185, color='blue', alpha=0.3) #ESTO ES PARA PONER LA SOMBRA
plt.hist(M_mujeres, bins=20, color="orange",label="media mujeres>18")
plt.axvspan(150, 170, color='darkorange', alpha=0.5) #ESTO ES PARA PONER LA SOMBRA
plt.title("Melvin Guaygua", loc="center", fontdict=font)
plt.legend()
plt.show()

fig, (ax1,ax2) = plt.subplots(1,2)
ax1.set_title("Melvin Guaygua", fontdict=font)
ax1.hist(M_hombres, bins=20)
ax1.axvspan(165, 185, color='blue', alpha=0.3) #ESTO ES PARA PONER LA SOMBRA
ax1.set(xlabel='media hombres > 18') #para poner un label a cada histograma
ax2.hist(M_mujeres, bins=20, color="orange")
ax2.axvspan(150, 170, color='darkorange', alpha=0.5) #ESTO ES PARA PONER LA SOMBRA
ax2.set(xlabel='media mujeres > 18') #para poner un label a cada histograma

ax1.vlines(x=mediaHombre, ymin=0, ymax=600, linestyles ="solid", colors ="cyan") #ESTO ES PARA PONER LA LINEA VERTICAL 
ax2.vlines(x=mediaMujeres, ymin=0, ymax=600, linestyles ="solid", colors ="cyan")
plt.show()
```

Haz la gr√°fica del analisis de frecuencias realizado del archivo "../04_RW/el_quijote.txt"

```python
x = np.array(["A","B","C","D","E","F","G","H","I","J","K","L","M","N","√ë","O","P","Q","R","S","T","U","V","W","X","Y","Z"])
#--
abc="ABCDEFGHIJKLMN√ëOPQRSTUVWXYZ"

fichero = open("el_quijote.txt",encoding="utf8")
contenido = fichero.read()

s = []
cont = 0
for i in contenido:
    s.append(i.upper().replace("√Å","A").replace("√â","E").replace("√ç","I").replace("√ì","O").replace("√ö","U"))
    if i.isalpha(): #esto era lo que me faltaba comprobar, comprobar que sea una letra
        cont+=1
f=[]
for i in abc:
    f.append(0)
    
for i in s:
    for j in range(27):
        if i==abc[j]:
            f[j]+=1

print("veces que se repite cada letra:")
print(f)
#--
y = np.array(f)

plt.bar(x,y)
plt.show()
```
