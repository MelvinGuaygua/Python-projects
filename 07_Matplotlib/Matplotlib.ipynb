### Scatter

Para el siguiente gráfico "scatter" utilizaremos un cmap, los cmap disponibles están en la siguiente página:  
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
import matplotlib.pyplot as plt
import numpy as np

#gráfico de burbujas o grafico de bolas
def fun(i,j):
    return (j**2)*(i**2)

x=[]
y=[]
z=[]
N=10
step=2
for i in np.arange(-N, N+step, step):
    for j in np.arange(-N, N+step, step):
        x.append(i)
        y.append(j)
        z.append(fun(i,j))

#puedo probar a borrar el size
# plt.scatter(x, y, c=z, cmap='viridis',s=8)
# plt.show()
plt.scatter(x, y, c=z, cmap='viridis',s=z, alpha=0.5) # x y z son arrays de numeros
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)
plt.show()
```
