```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 1) 
y = x*x

plt.plot(x,y,marker = '*', linestyle = ':',linewidth = '2.5',color='g',label='$x^2$')
plt.plot(x,x*8,marker = '+', linestyle = '--',linewidth = '1.5',color='r',label='8*x')

font = {'family':'serif','color':'darkred','size':15}
plt.title("Título", loc = 'center', fontdict = font)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = np.arange(0, 10, 1) 
y = x*x
# plt.plot(x,y)  #cada vez que haces plot estas pintanto una grafica

fig, ax = plt.subplots()
plt.plot(x,y)
ax.axvspan(5, 10, alpha=0.3, color='blue')

plt.show()
```
