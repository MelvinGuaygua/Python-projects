```python
#EJER_1 V1.0 HECHO IMPRIMIENDO CADA DIGITO
objeto = open("archivo.txt", "r")
comprobar=1
for i in range(16): #va a leer caracter por caracter, 4 veces
    caracter = objeto.read(1) # imprime caracter por caracter (salto de linea=caracter)
    if caracter.isdigit(): #con esto compruebo si es un numero
        caracter = int(caracter) #sin esta linea caracter sigue siendo un string y me da error al .format(caracter)
        if comprobar%2!=0: #si comprobar es impar se imprime 1 decimal
            s2="   "+"{:.1f}".format(caracter) #con esto agrego identacion al principio 
            print(s2, end="   ") #elimino el salto de linea cuando comprobar es impar
            comprobar+=1
        elif comprobar%2==0: #si comprobar no es par se imprimen 2 decimales
            print("{:.2f}".format(caracter)+"   ")
            comprobar+=1

objeto.close()
```

```python
#LO QUE ME HA DICHO EL PROFE EJER_4
archivo = "matriz.txt"
text = open(archivo).readlines()

print("imprimiendo todas las lineas: ") #imprimir todas las lineas
for line in text:
    print(line.split())

x=[]
for line in text:
    x.append(line.split())

print("lista primera fila: ")
print(x[0])
print("numero especifico: ")
print(x[0][0]) #[fila][columna]
print(x[0][1])
print(x[0][2])
print("lista segunda fila: ")
print(x[1])
print("numero especifico: ")
print(x[1][0])
print(x[1][1])
print(x[1][2])

#AQUI COMIENZA MI MATRIZ
print("parte A: ")
print(x[0][0])
print(x[1][1])
print(x[2][2])
var1=x[0][0]
var2=x[1][1]
var3=x[2][2]
print("resultado:")
resultado1=int(var1)*int(var2)*int(var3)
print(resultado1)

print("parte A: ")
print(x[1][0])
print(x[2][1])
print(x[0][2])
var4=x[1][0]
var5=x[2][1]
var6=x[0][2]
print("resultado:")
resultado2=int(var4)*int(var5)*int(var6)
print(resultado2)

print("parte A:")
print(x[2][0])
print(x[0][1])
print(x[1][2])
var7=x[2][0]
var8=x[0][1]
var9=x[1][2]
print("resultado:")
resultado3=int(var7)*int(var8)*int(var9)
print(resultado3)

print("parte B:")
print(x[2][0])
print(x[1][1])
print(x[0][2])
var10=x[2][0]
var11=x[1][1]
var12=x[0][2]
print("resultado:")
resultado4=int(var10)*int(var11)*int(var12)
print(resultado4)

print("parte B: ")
print(x[0][1])
print(x[1][0])
print(x[2][2])
var13=x[0][1]
var14=x[1][0]
var15=x[2][2]
print("resultado:")
resultado5=int(var13)*int(var14)*int(var15)
print(resultado5)

print("parte B:")
print(x[0][0])
print(x[1][2])
print(x[2][1])
var16=x[0][0]
var17=x[1][2]
var18=x[2][1]
print("resultado:")
resultado6=int(var16)*int(var17)*int(var18)
print(resultado6)

resultadoFinal=resultado1+resultado2+resultado3-(resultado4+resultado5+resultado6)
print("\nResultado de la matriz: "+str(resultadoFinal))
```
