```python
from datetime import date

class Persona:
    #para crear atributos y metodos privados tengo que poner __
    __pensamiento = "hola mundo"
    cara = "alegre"
    
    #para crear el constructor se crea la funcion __init__ (sirve para inicializar productos)
    def __init__(self, nombre="melvin", año_nacimiento=1999):
        self.nombre = nombre
        self.año_nacimiento = año_nacimiento
        self.edad = self.get_edad() #HAY QUE PONER LOS PARENTESIS
    
    #metodo para calcular la edad
    def get_edad(self):  
        return date.today().year - self.año_nacimiento
    
    def info(self):
        print("*nombre "+self.nombre)
        print("*año nacimiento: "+str(self.año_nacimiento))
        print("*edad: "+str(self.edad))
        
    def get_pensamiento(self):
        print(self.__pensamiento)
        
    def diferencia_edad(self, otra_persona):
        print("diferencia de edad "+str(abs(self.edad-otra_persona.edad)))
    
    def diferencia_edad2(self,otra_persona): 
        print("Me llevo "+str(abs(self.edad-otra_persona.edad))+" años con "+otra_persona.nombre)
    

persona1 = Persona() #inicializo a la persona
persona1.info() #saco informacion de la persona
```

```python
persona2 = Persona("pepe",1980)
persona2.info()

persona1.diferencia_edad(persona2)
persona1.diferencia_edad2(persona2)
```

```python
class Cuenta:
#     def __init__(self):
#         pass
    
    def __init__(self, ti=None, can=0.0):
        self.__titular = ti
        self.__cantidad=can
    
#     def __init__(self, ti, can=0.0):
#         self.titular = ti
#         self.cantidad=can
        
    #getters
    def get_titular(self):
        return self.__titular
    def get_cantidad(self):
        return self._cantidad
    
    #setters
    def set_titular(self, titular):
        self.__titular = titular
    def set_cantidad(self, cantidad):
        self.__cantidad = cantidad
        
    #toString
    def mostrar(self):
        print("-----------------")
        print("titular: "+str(self.__titular))
        print("cantidad: "+str(self.__cantidad))
        
    def ingresar(self,cantidad):
        if cantidad<0: 
            print("no se puede ingresar numeros negativos")
        else:
            saldoNuevo = self.__cantidad + cantidad
            self.set_cantidad(saldoNuevo) 
            print("saldo despues de ingresar: "+str(self.__cantidad))
            
    def retirar(self, cantidad):
        saldoNuevo = self.__cantidad - cantidad
        self.set_cantidad(saldoNuevo)
        print("saldo despues de retirar: "+str(self.__cantidad))
        

cuenta1 = Cuenta("melvin")
cuenta1.mostrar()
cuenta1.set_titular("eee")
cuenta1.mostrar()

cuenta2 = Cuenta("pepe",12)
cuenta2.mostrar()

cuenta2.ingresar(12)
cuenta2.retirar(4)
```

```python
#EJER 3 TERMINADO
class Persona_tlf:
    def __init__(self, fecha, genero, nombre, apellido, tel):
        self.__fecha = fecha
        self.__genero = genero
        self.__nombre = nombre
        self.__apellido = apellido
        self.__tel = tel
        
    def info(self):
        print("--INFO--")
        print("fecha: "+str(self.__fecha))
        print("genero: "+self.__genero)
        print("nombre: "+self.__nombre)
        print("apellido: "+self.__apellido)
        print("tel: "+self.__tel)

    def tlf_info(self):
        if self.__tel[0] == "9" or self.__tel[0] == "8":
            print("telefono fijo: "+self.__tel)
        elif self.__tel[0] == "6" or self.__tel[0] == "7":
            print("telefono movil: "+self.__tel)
        
    def get_tlf(self):
        print(self.__tel)
    
    def set_tlf(self, tel):
        self.__tel = tel
        
    def cambiar_tlf(self, telNuevo):
        telString=str(telNuevo)
        if len(telString)!= 9:
            print("->numero NO cambiado, necesitamos 9 cifras")
        else: 
            self.set_tlf(telString)
            print("->numero cambiado, ahora es: "+self.__tel)
        

#descomenta la siguientes 2 lineas
persona_tlf01 = Persona_tlf(1979,"Hombre","Manuel","López","987654321")
persona_tlf01.get_tlf()

#salida:
#987654321

#descomenta la siguiente linea
persona_tlf01.tlf_info()

#salida:
#Mi teléfono es un fíjo con número 987654321

# descomenta la siguiente linea
persona_tlf01.cambiar_tlf(1234)
persona_tlf01.info()
#salida:
#Número no cambiado, necesitamos un número de 9 cifras

#descomenta la siguiente linea
persona_tlf01.cambiar_tlf(654987321)
persona_tlf01.info()

#salida:
#Número cambiado, Mi teléfono es un móvil con número 987654321
```
