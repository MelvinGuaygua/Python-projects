```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Soleado", "Nublado", "Lluvia", "Nieve"]) #con este establezco
y = np.array([70, 20, 9, 1])

plt.bar(x,y, color="red")
plt.show()

plt.barh(x,y, height=0.1)
plt.show()

plt.pie(y, labels=x)
plt.show()
```
