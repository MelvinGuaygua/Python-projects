Vamos a hacer una difución del calor utilizando solo los 4 primero vecinos, es decir:

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph([(1, 2),(1, 2),(1, 3),(1, 4),(1, 5)])
L=1.0
pos = {1: (0, 0), 2: (L, 0), 3: (-L, 0), 4: (0,L), 5: (0,-L)}
options = {"font_size": 12,"node_size": 600,"node_color": "white","edgecolors": "black","linewidths": 3,"width": 3,}
nx.draw_networkx(G, pos, **options)
plt.axis("off")
plt.show()
```

```python
A = np.zeros([50,50])
A[int(len(A)/2),int(len(A)/2)] = 100

#Difundimos el calor
for t in range(100):
    for i in range(1,len(A)-1):
        for j in range(1,len(A)-1):
            A[i][j]=(A[i+1][j]+A[i-1][j]+A[i][j+1]+A[i][j-1])/4
            A[int(len(A)/2),int(len(A)/2)] = 100

sns.heatmap(A, cmap = 'Reds', yticklabels=False, xticklabels=False)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = np.arange(len(A))
Y = np.arange(len(A))
X, Y = np.meshgrid(X, Y)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

surf = ax.plot_surface(X, Y, A, cmap='Reds',linewidth=0)
 
```

```python
"""Esribe tu código aquí"""

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set_style('darkgrid') # darkgrid, whitegrid, dark, white, ticks
iris = sns.load_dataset('iris')
iris.head()

df_setosa = iris.loc[iris['species'] == "setosa"]
# display(df_setosa.head())

df_versicolor = iris.loc[iris['species'] == "versicolor"]
# display(df_versicolor.head())

df_virginica = iris.loc[iris['species'] == "virginica"]
# display(df_virginica.head())

fig, axes = plt.subplots(ncols=3, figsize=(16, 4))
ax1, ax2, ax3 = axes

g1 = sns.heatmap(df_setosa.corr(), annot = True, cmap = 'coolwarm',ax=ax1)
g2 = sns.heatmap(df_versicolor.corr(), annot = True, cmap = 'coolwarm',ax=ax2)
g3 = sns.heatmap(df_virginica.corr(), annot = True, cmap = 'coolwarm',ax=ax3)

nombre=["setosa","versicolor","virginica"]
cont=0
for ax in [g1,g2,g3]:
    ax.set_title(nombre[cont]) #para setear el titulo con valores de la lista
    cont+=1
    tl = ax.get_xticklabels() #para obtener los xticklabels de cada ax
    ax.set_xticklabels(tl, rotation=90) #para rotar los xticklabels
    
# para quitar los yticks del segundo y tercero:
g2.set_yticks([])
g3.set_yticks([])
    
plt.show()
```

¿Se pueden agrupar con clustermap()?    
En caso de que se pueda ¿Qué interpretación puedes hacer de esta agrupación?

```python
"""Esribe tu código aquí"""

import investpy
lista=['ACX', 'BBVA', 'BKT', 'BME', 'ANA', 'CABK', 'ENAG', 'FCC', 'NTGY', 'SGREN', 'GRLS', 'FER']

df_ibex35 = pd.DataFrame()

for i in lista:
    df = investpy.get_stock_historical_data(stock=i,country='spain',from_date='01/01/2015',to_date='01/01/2020')
    df[i]=(df['Close']+df['Open'])/2
    df_ibex35[i] = pd.DataFrame(df[i])

df_ibex35.head()

sns.heatmap(df_ibex35.corr(), annot = True, cmap = 'coolwarm')
```
